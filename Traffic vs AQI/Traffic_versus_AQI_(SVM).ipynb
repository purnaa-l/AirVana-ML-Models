```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR


# Load dataset
df = pd.read_csv("realistic_indian_aqi_prediction_data.csv")

df.head()


# # Select features and target
# features = ['temperature_celsius', 'wind_kph', 'pressure_mb', 'humidity', 'cloud', 'uv_index']
# target = 'air_quality_PM2.5'

```

```python
features = ['Vehicle Count', 'Average Speed (km/h)', 'Flow Rate (vehicles/hour)', 'Vehicle Density (vehicles/km)']
target = 'AQI'
```

```python
# Drop rows with missing values in selected columns
df = df[features + [target]].dropna()

# Feature matrix and target vector
X = df[features]
y = df[target]

# Scale features (optional but good for some models)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor=svr = SVR(kernel='poly')
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R² Score:", r2)
print("RMSE:", rmse)
```
